```python
def grid(rmin,rmax,nx,spacing):
    if spacing == "log":
        
        f   =  1/(nx-1)*np.log(rmax/rmin)
        d_r =  rmin/2*(1-1/np.exp(f))

        rc = (rmin    )*np.exp(f*(np.arange(nx  ))) #arange -> 0 to nx-1
        re = (rmin-d_r)*np.exp(f*(np.arange(nx+1))) #arange -> 0 to nx
        
    elif spacing == "lin":
        rc  = np.linspace(rmin,rmax,nx)
        dr  = rc[1]-rc[0]
        re  = np.linspace(rmin-dr/2,rmax+dr/2,nx+1)
    
    return np.asarray(re),np.asarray(rc)
```

```python
#nx         = 500
nx_l = [800]#,400,600,800]
Mdisk_perc = 0.3
m          = 4

#rin        = 0.25
#rin_l = [0.25,0.20,0.15,0.10]
rin_l = [0.2]

rD         = 1.0
#spacing    = "log"
spacing_l = ["lin","log"]
G          = 1.0
K          = 0.25
M          = 0.6
gammap     = 2.0
f = np.copy(gammap)
b = 10

profile = "bump"
#profile = "power"
    
for spacing in spacing_l:
    for nx in nx_l:
        for rin in rin_l:
            Md         = (Mdisk_perc*M)/(1.0-Mdisk_perc)
            re,r       = grid(rin,rD,nx,spacing)
            
            extention = profile+"_"+str(nx)+"_"+str(int(rin*100))+"_"+str(int(rD*100))+"_"+spacing+"_"+str(m)+"_"+str(int(Mdisk_perc*100))+"_"+str(int(gammap*10))+"_"+str(b)

            sigma0,omega,kappa,cs = np.loadtxt("./files/quantity/"+extention+".txt",np.asarray([sigma0,omega,kappa,cs]).T)
            
            beta_RR_list = np.loadtxt("./files/roots/"+extention+".txt")
            beta_RR = beta_RR_list[:,0] + 1j*beta_RR_list[:,1]
            
            sigma1_list = np.loadtxt("files/sigma1/"+extention+".txt")
            
            sigma1 = sigma1_list[:,0] + 1j*sigma1_list[:,1]
            
            plt.plot(r,np.abs(sigma1),label=str(nx)+" "+str(rin))
            plt.xscale("log")
            plt.yscale("log")
plt.legend(frameon=False)
```
